Описание для файла MEbatch_hs

hs (hinode-style): Единицы приведены к используемым в файлах Hinode Level-2

ME - функция, возвращающая значение профиля в модели Милна-Эддингтона с заданными параметрами на заданной длине волны
    

ME_ff - аналогично предыдущей, но с добавлением двух вариативных параметров - фактора заполнения и доплеровского сдвига невозмущенного потока

Фактор заполнения (filling factor) - приближение, в котором считается, что часть пикселя содержит элемент магнитного поля (B != 0), другая же часть невозмущена (B = 0). 

line_vec: набор параметров линии, известных до проведения инверсии
    wl0 - длина волны в ангстремах (константно 6302.5)
    g - фактор Ланде (константно 2.5)
    mu - угол между нормалью к поверхности и лучом зрения (константно 1, см. my.ipynb)
    
param_vec: набор параметров, которые мы ищем в инверсии ("вариативные" параметры)
    B - модуль магнитного поля (параметр 1 в Hinode Level 2) - гауссы
    theta - наклон магнитоного поля (параметр 2) - градусы
    xi - азимут магнитного поля (3) - градусы
    D - доплеровская ширина профиля (6) - миллиангстремы
    gamma - лоренцовская ширина профиля (8) - в единицах D
    etta_0 - сила линии (7)
    S_0 - функция источников (9)
    S_1 - декремент функции источников (10)
    Dop_shift - доплеровский сдвиг (5) - км/с
    
дополнительные параметры для ME_ff:
    filling_factor - фактор заполнения (12)
    stray_shift - доплеровский сдвиг невозмущенного компонента (13) - км/с

x: длина волны (или массив), в которой требуется определить значение функции
